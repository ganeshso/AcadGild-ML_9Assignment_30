# Load Libraries

```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import matplotlib.pyplot as plt
```

```python
img = misc.face(gray = True)
plt.imshow(img)
```

```python
img.shape
```

# Convert the Image into an Array

```python
img_r = (img / 255.0).reshape(-1,3)

img_r.shape
```

**Use K-Means for 5 Clusters from Reshaped Image Array**

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=5).fit(img_r)
y_pred_kmeans = k_means.predict(img_r)
```

```python
# how many centriods 
k_means.cluster_centers_.shape
```

```python
k_means.labels_.shape
```

**Create New Image Array from the Generated Clusters and Labels**

```python
new_img = k_means.cluster_centers_[k_means.labels_]

new_img.shape
```

**Reshape the New Image Array with Original Image Dimensions**

```python
new_img = np.reshape(new_img, (img.shape))

new_img.shape
```

**Plotting Original and Color Compressed Image**

```python
fig = plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(img)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-Means')
ax.imshow(new_img)
plt.show()
```
